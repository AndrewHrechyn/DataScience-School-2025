# Частина 2. Підготовка даних та створення нових ознак

1) Створення метрики "Вартість однієї покупки"

```python
df['Avg Spend per Item'] = df.apply(lambda row:
                                    row['Total Spend'] / row['Items Purchased']
                                    if row['Items Purchased'] != 0 else 0,
                                    axis = 1
                                     )
```

```python
df.head(5)
```

2) Створення сегментів за активністю.

```python
df['Active Status'] = df.apply(lambda x: 'Active' if x['Days Since Last Purchase'] > 30 else 'Inactive', axis=1)
```

```python
df.head(5)
```

Питання 2: Як знижки впливають на поведінку клієнтів?

```python
discount_analysis = df.groupby('Discount Applied').agg({
    'Total Spend': 'mean',
    'Items Purchased': 'mean'
}).reset_index()
```

```python
spend_with_discount = discount_analysis.loc[discount_analysis['Discount Applied'] == True, 'Total Spend'].values[0]
spend_without_discount = discount_analysis.loc[discount_analysis['Discount Applied'] == False, 'Total Spend'].values[0]
items_with_discount = discount_analysis.loc[discount_analysis['Discount Applied'] == True, 'Items Purchased'].values[0]
items_without_discount = discount_analysis.loc[discount_analysis['Discount Applied'] == False, 'Items Purchased'].values[0]
```

```python
if spend_with_discount > spend_without_discount:
    print(f"Клієнти зі знижками витрачають більше (в середньому {spend_with_discount:.2f} проти {spend_without_discount:.2f}).")
else:
    print(f"Клієнти без знижок витрачають більше (в середньому {spend_without_discount:.2f} проти {spend_with_discount:.2f}).")

if items_with_discount > items_without_discount:
    print(f"Клієнти зі знижками купують більше товарів (в середньому {items_with_discount:.2f} проти {items_without_discount:.2f}).")
else:
    print(f"Клієнти без знижок купують більше товарів (в середньому {items_without_discount:.2f} проти {items_with_discount:.2f}).")
```

Питання 4: Чи дійсно жінки з Gold-членством мають вищий рівень задоволеності, ніж чоловіки з тим же статусом?

# Частина 5. Бонусні завдання (додаткові бали)


Бонус 1: Розширений аналіз сегментації.

```python
avg_spend = df['Total Spend'].mean()
```

```python
vip = df[(df['Membership Type'] == 'Gold') & (df['Total Spend'] > avg_spend)]
```

```python
non_vip = df[~((df['Membership Type'] == 'Gold') & (df['Total Spend'] > avg_spend))]
```

```python
print("VIP рейтинг: ", vip['Average Rating'].mean())
print('VIP задоволенність: ', vip['Satisfaction Level'].value_counts(normalize=True) * 100)

print("VIP рейтинг інших клієнтів: ", non_vip['Average Rating'].mean())
print('VIP задоволенність інших клієнтів: ', non_vip['Satisfaction Level'].value_counts(normalize=True) * 100)
```
